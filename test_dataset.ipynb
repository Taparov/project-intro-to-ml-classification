```python
import kaggle
from sklearn.preprocessing import OrdinalEncoder, StandardScaler, LabelEncoder
from sklearn.feature_selection import mutual_info_classif
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('./dataset/UCI_Credit_Card.csv')
df.dropna(inplace=True)
df.head(5)
```

```python
df['SEX'] = np.where(df['SEX'] == 1, 0, 1)
```

```python
df['MARRIAGE'].unique()
```

```python
# drow rows where Education is 0, 4,5, or 6
df = df[~df['EDUCATION'].isin([0, 4, 5, 6])]
#  drop rows where Mariage is 0
df = df[~df['MARRIAGE'].isin([0])]
# drop roes where PAY_0,2,3,4,5,6 is -2
df = df[~df['PAY_0'].isin([-2])]
df = df[~df['PAY_2'].isin([-2])]
df = df[~df['PAY_3'].isin([-2])]
df = df[~df['PAY_4'].isin([-2])]
df = df[~df['PAY_5'].isin([-2])]
df = df[~df['PAY_6'].isin([-2])]
# change valuse in pay from -1 to 0
df['PAY_0'] = np.where(df['PAY_0'] == -1, 0, df['PAY_0'])
df['PAY_2'] = np.where(df['PAY_2'] == -1, 0, df['PAY_2'])
df['PAY_3'] = np.where(df['PAY_3'] == -1, 0, df['PAY_3'])
df['PAY_4'] = np.where(df['PAY_4'] == -1, 0, df['PAY_4'])
df['PAY_5'] = np.where(df['PAY_5'] == -1, 0, df['PAY_5'])
df['PAY_6'] = np.where(df['PAY_6'] == -1, 0, df['PAY_6'])
# shift and reverse valus of MARRIAGE down by 1
df['MARRIAGE'] = np.where(df['MARRIAGE'] == 1, 0, 1)
# shift Education values down by 1
df['EDUCATION'] = df['EDUCATION'] - 1

```

```python
# drop ID column
df.drop(columns=['ID'], inplace=True)
```

```python
df['EDUCATION'].unique()
```

```python
df.head(5)
```

```python
# Classification Mutual Information
discrete_features = [False] * 23
for col in [1,2,3,5,6,7,8,9,10]:
    discrete_features[col] = True
X = df.drop(columns=['default.payment.next.month'])
y = df['default.payment.next.month']
mi = mutual_info_classif(X, y, discrete_features=discrete_features)
mi_scores = pd.Series(mi, index=X.columns).sort_values(ascending=False)
print(mi_scores)
```

```python
# Regression Mutual Information
from sklearn.feature_selection import mutual_info_regression
X = df.drop(columns=['LIMIT_BAL'])
y = df['LIMIT_BAL']
discrete_features = [False] * 23
for col in [1,2,3,5,6,7,8,9,10,23]:
    discrete_features[col-1] = True

mi_scores = mutual_info_regression(X, y, discrete_features=discrete_features)
mi_df = (
    pd.Series(mi_scores, index=X.columns)
    .sort_values(ascending=False)
)

print("Mutual Information scores:")
print(mi_df)
```
